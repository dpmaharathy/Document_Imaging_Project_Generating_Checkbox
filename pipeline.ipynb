```python
import os
import re

# Define input and output directories
input_dir = r'C:\Users\HP\Downloads\Forms_updated\full_latex_files'
output_dir = r'C:\Users\HP\Downloads\Forms_updated\single_bounded_latex_files'

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Define the regex pattern to search for \CheckBox with optional attributes
pattern = r'\\CheckBox\['

# Function to replace each occurrence of \CheckBox with modified attributes
def replace_checkbox(match):
    return f'\\CheckBox[bordercolor = 1 1 1, '

# Process LaTeX files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.tex'):
        input_file = os.path.join(input_dir, filename)
        output_folder = os.path.join(output_dir, filename[:-4])  # Create folder based on original filename
        os.makedirs(output_folder, exist_ok=True)  # Ensure output folder exists

        # Read the LaTeX file
        with open(input_file, 'r') as f:
            latex_code = f.read()

        # Replace \CheckBox in the LaTeX code
        modified_latex_code_all = re.sub(pattern, replace_checkbox, latex_code)

        # Write modified LaTeX code to the output file (all)
        output_file_all = os.path.join(output_folder, f'{filename[:-4]}_all.tex')
        with open(output_file_all, 'w') as f:
            f.write(modified_latex_code_all)

        # Find all occurrences of \CheckBox and create separate files for each occurrence
        checkbox_occurrences = re.finditer(pattern, modified_latex_code_all)
        inserted_text = 'bordercolor = 0 0 0, '
        for i, match in enumerate(checkbox_occurrences, start=1):
            start, end = match.span()
            modified_latex_code_i = modified_latex_code_all[:start+10] + inserted_text + modified_latex_code_all[start+30:]
            output_file_i = os.path.join(output_folder, f'{filename[:-4]}_{i}.tex')
            with open(output_file_i, 'w') as f:
                f.write(modified_latex_code_i)

            print(f'File {output_file_i} generated in folder {filename[:-4]}.')

print('All LaTeX files processed.')
```

```python
import os
import numpy as np
from PIL import Image
import fitz  # PyMuPDF

def pdf_to_jpg(pdf_file, quality=200, resolution=500):
    images = []
    pdf_document = fitz.open(pdf_file)
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        zoom_matrix = fitz.Matrix(resolution/72.0, resolution/72.0)  # Set zoom for high resolution
        pix = page.get_pixmap(matrix=zoom_matrix)  # Render page at high resolution
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        images.append(img)
        img.save(pdf_file[:-4] + '_pdf_to_jpg_' + str(page_num) + '_.jpg', quality=quality)  # Set quality parameter
    pdf_document.close()
    return images

def subtract_images(image_file, reference_file, output_file):
    # Convert PDFs to images (JPG format)
    image_images = pdf_to_jpg(image_file)
    reference_images = pdf_to_jpg(reference_file)
    
    # Convert images to numpy arrays
    image_array = np.array(image_images[0])
    reference_array = np.array(reference_images[0])

    # Ensure both images have the same size
    if image_array.shape != reference_array.shape:
        raise ValueError("Image and reference image must have the same dimensions.")

    # Subtract the reference image from the image
    result_array = image_array - reference_array

    # Invert colors to make the foreground black and background white
    result_array = -result_array
    result_array = 255 - result_array

    # Form a new image
    result_image = Image.fromarray(result_array.astype('uint8'), 'RGB')

    # Save the resulting image
    result_image.save(output_file)

# Input and output directories
input_directory = r'C:\Users\HP\Downloads\Forms_updated\pdfs'
output_directory = r'C:\Users\HP\Downloads\Forms_updated\jpgs'

# Iterate over each folder in the input directory
for folder_name in os.listdir(input_directory):
    folder_path = os.path.join(input_directory, folder_name)
    if os.path.isdir(folder_path):
        output_folder_path = os.path.join(output_directory, folder_name)
        os.makedirs(output_folder_path, exist_ok=True)

        # Find the image files and the reference file
        image_files = []
        reference_file = None
        for filename in os.listdir(folder_path):
            if filename.endswith('.pdf'):
                if filename.endswith('_all.pdf'):
                    reference_file = os.path.join(folder_path, filename)
                else:
                    image_files.append(os.path.join(folder_path, filename))

        # Process each image file with the reference file
        for image_file in image_files:

            output_file = os.path.join(output_folder_path, os.path.basename(image_file)[:-4] + '.jpg')
            print(image_file)
            subtract_images(image_file, reference_file, output_file)
            print(f'Resulting image saved to {output_file}')
```

```python
import os
import cv2
import numpy as np

def extract_boundary_edges(image_folder, output_directory):
    # Iterate over each folder in the image directory
    for folder_name in os.listdir(image_folder):
        folder_path = os.path.join(image_folder, folder_name)
        if os.path.isdir(folder_path):
            output_file = os.path.join(output_directory, f'{folder_name}.txt')

            # Open output file for writing YOLO format coordinates
            with open(output_file, 'w') as f:
                # Iterate over each image file in the folder
                for filename in os.listdir(folder_path):
                    if filename.endswith('.jpg'):
                        # Load image, grayscale, median blur, sharpen image
                        image_path = os.path.join(folder_path, filename)
                        image = cv2.imread(image_path)
                        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                        blur = cv2.medianBlur(gray, 5)
                        sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
                        sharpen = cv2.filter2D(blur, -1, sharpen_kernel)

                        # Threshold and morph close
                        thresh = cv2.threshold(sharpen, 125, 255, cv2.THRESH_BINARY_INV)[1]
                        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
                        close = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=2)

                        # Find contours and filter using threshold area
                        cnts = cv2.findContours(close, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                        cnts = cnts[0] if len(cnts) == 2 else cnts[1]

                        # Get image width and height
                        image_width, image_height = image.shape[1], image.shape[0]

                        # Iterate over each contour and write YOLO format coordinates to file
                        for c in cnts:
                            x, y, w, h = cv2.boundingRect(c)
                            # Normalize coordinates and dimensions
                            x_center = (x + w / 2) / image_width
                            y_center = (y + h / 2) / image_height
                            width_normalized = w / image_width
                            height_normalized = h / image_height
                            # Write YOLO format coordinates to file
                            f.write(f'0 {x_center:.6f} {y_center:.6f} {width_normalized:.6f} {height_normalized:.6f}\n')

# Input and output directories
image_directory = r'C:\Users\HP\Downloads\Forms_updated\jpgs'
output_directory = r'C:\Users\HP\Downloads\Forms_updated\annotations'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# Extract boundary edges and write YOLO format coordinates for each folder in the input directory
extract_boundary_edges(image_directory, output_directory)
```

```python
import os
import fitz  # PyMuPDF
from PIL import Image


def pdf_to_jpg_single(pdf_file, output_dir, quality=95, resolution=500):
    # Open the PDF file
    pdf_document = fitz.open(pdf_file)

    # Load the first page
    page = pdf_document.load_page(0)
    
    # Render the page as an image
    zoom_matrix = fitz.Matrix(resolution/72.0, resolution/72.0)  # Set zoom for high resolution
    pix = page.get_pixmap(matrix=zoom_matrix)  # Render page at high resolution
    img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

    # Construct the output file path
    output_file = os.path.join(output_dir, f'{os.path.basename(pdf_file)[:-4]}.jpg')

    # Save the image
    img.save(output_file, format='JPEG', quality=quality)
    print(f'Image saved: {output_file}')

    # Close the PDF document
    pdf_document.close()

    return output_file


# Directory containing PDF files
pdf_directory = r'C:\Users\HP\Downloads\Forms_updated\full_pdfs'

# Directory to save JPEG files
output_directory = r'C:\Users\HP\Downloads\Forms_updated\full_jpgs'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# Iterate over each file in the PDF directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        pdf_file_path = os.path.join(pdf_directory, filename)
        
        # Convert PDF to JPEG
        print(pdf_file_path)
        pdf_to_jpg_single(pdf_file_path, output_directory)
```

```python
import os
import cv2

# Directory containing annotation files
annotations_directory = r'C:\Users\HP\Downloads\Forms_updated\annotations'

# Directory containing image files
images_directory = r'C:\Users\HP\Downloads\Forms_updated\full_jpgs'

# Directory to save annotated images
output_directory = r'C:\Users\HP\Downloads\Forms_updated\annotated_images'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# Iterate over each annotation file
for annotation_file in os.listdir(annotations_directory):
    if annotation_file.endswith('.txt'):
        annotation_path = os.path.join(annotations_directory, annotation_file)
        image_path = os.path.join(images_directory, f'{os.path.splitext(annotation_file)[0]}.jpg')
        print(image_path)
        # Read the image
        image = cv2.imread(image_path)

        # Read the annotations
        with open(annotation_path, 'r') as f:
            lines = f.readlines()

        # Parse annotations and draw bounding boxes
        for line in lines:
            # Split the line into label and bounding box coordinates
            label, *box = line.strip().split()
            box = list(map(float, box))  # Convert coordinates to floats

            # Convert normalized coordinates to absolute coordinates
            # print(image)
            height, width, _ = image.shape
            x_center, y_center, w, h = box
            x1 = int((x_center - w/2) * width)
            y1 = int((y_center - h/2) * height)
            x2 = int((x_center + w/2) * width)
            y2 = int((y_center + h/2) * height)

            # Draw the bounding box on the image
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Save the annotated image
        output_image_path = os.path.join(output_directory, f'{os.path.splitext(annotation_file)[0]}_annotated.jpg')
        cv2.imwrite(output_image_path, image)
        print(f'Annotated image saved: {output_image_path}')
```
