```python
import fitz  # PyMuPDF
from PIL import Image
import os

def pdf_to_images(pdf_file):
    images = []
    pdf_document = fitz.open(pdf_file)
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        images.append(img)
    pdf_document.close()
    return images

def save_images_as_pdf(images, output_file):
    images[0].save(output_file, save_all=True, append_images=images[1:])

def subtract_pdf_from_images(pdf_file, other_pdf_files, output_files):
    main_images = pdf_to_images(pdf_file)
    other_images = [pdf_to_images(pdf) for pdf in other_pdf_files]

    for i in range(len(main_images)):
        for other_image in other_images:
            main_images[i] = Image.alpha_composite(main_images[i], other_image[i])

    for output_file, image in zip(output_files, main_images):
        image.save(output_file)

# Example usage
pdf_file = "C:\Users\HP\Downloads\Forms\forms.pdf"
other_pdf_files = ["C:\Users\HP\Downloads\Forms\forms (1).pdf", "C:\Users\HP\Downloads\Forms\forms (2).pdf", "C:\Users\HP\Downloads\Forms\forms (3).pdf"]
output_files = ["result1.pdf", "result2.pdf", "result3.pdf"]

subtract_pdf_from_images(pdf_file, other_pdf_files, output_files)
```
