```python
import os
import itertools

def generate_combinations(files):
    combinations = []
    for r in range(1, len(files) + 1):
        combinations.extend(itertools.combinations(files, r))
    return combinations

def insert_checkboxes(latex_file):
    with open(latex_file, 'r') as f:
        content = f.readlines()

    checkboxes = ['\\item[\\CheckBox] Checkbox Item\n' for _ in range(5)]  # Insert 5 checkboxes

    with open(latex_file, 'w') as f:
        for line in content:
            if line.strip() == '% INSERT_CHECKBOXES_HERE':
                f.writelines(checkboxes)
            else:
                f.write(line)

def save_modified_file(original_file_path, output_file_path):
    print(f"Copying {original_file_path} to {output_file_path}")
    with open(original_file_path, 'r') as original_file:
        with open(output_file_path, 'w') as output_file:
            output_file.write(original_file.read())
    print(f"Inserting checkboxes into {output_file_path}")
    insert_checkboxes(output_file_path)

if __name__ == "__main__":
    files = [r'C:\Users\HP\Downloads\Forms\full_latex_files\f1.tex', r'C:\Users\HP\Downloads\Forms\full_latex_files\f2.tex',
             r'C:\Users\HP\Downloads\Forms\full_latex_files\f3.tex', r'C:\Users\HP\Downloads\Forms\full_latex_files\f4.tex']

    combinations = generate_combinations(files)

    for idx, combination in enumerate(combinations):
        output_dir = f'output_combination_{idx}'
        os.makedirs(output_dir, exist_ok=True)
        for file in combination:
            original_file_path = os.path.join('latex_files', file)
            output_file_path = os.path.join(output_dir, file)
            save_modified_file(original_file_path, output_file_path)
```
